```python
"""
MODIFICADO DE pyredatam.py en https://github.com/abenassi/pyredatam/blob/master/pyredatam/pyredatam.py
"""

import urllib.parse

def query_final(tipo=None,variables1=None, variables2=None,selection=None,area_break=None,
           universe_filter=None, title=None):
    head = "https://censos2017.inei.gob.pe/bininei/RpWebStats.exe/CmdSet?MAIN=WebServerMain.inl&BASE=CPV2017DI&LANG=esp&CODIGO=XXUSUARIOXX&ITEM=PROGRED&MODE=RUN&CMDSET="
    submit = "&Submit=Ejecutar"
    
    if tipo=="frequency": # Frequency
        return head+urllib.parse.quote(frequency_query(variables1,selection,area_break,universe_filter,title))+submit
    
    if tipo=="crosstab": # Crosstab
        return head+urllib.parse.quote(crosstab_query(variables1,variables2,selection,area_break,universe_filter,title))+submit
    
    else:
        return "No seleccion√≥ tipo de consulta"
# funcion para escribir consulta redatam en lenguaje redatam
def frequency_query(variables1,selection,area_break,
                   universe_filter, title):
    # RUNDEF section
    lines = _build_rundef_section(selection,universe_filter)
    # TABLE section
    lines.append("TABLE TABLE1")
#     lines.extend(_build_title(title))
    lines.append("    AS FREQUENCY")
    lines.append(_build_area_break(area_break))
    lines.append(_build_of_variables(variables1))

    return "\n".join(lines)

def crosstab_query(variables1,variables2,selection,area_break,
                   universe_filter,title):
    # RUNDEF section
    lines = _build_rundef_section(selection,universe_filter)

    # TABLE section
    lines.append("TABLE TABLE1")
    lines.extend(_build_title(title))
    lines.append(_build_of_variables2(variables1,variables2))
    lines.append(_build_area_break(area_break))

    return "\n".join(lines)

# funcion para escribir parametros de la primera linea de la consulta
def _build_rundef_section(selection=None, universe_filter=None):
    lines = ["RUNDEF Job"]
    lines.append(_build_selection_inline(selection))
    lines.append(_build_universe_filter(universe_filter))
    lines.append("")
    return lines

# funcion para escribir parametros de filtro universo
def _build_universe_filter(universe_filter):
    return "    UNIVERSE " + universe_filter if universe_filter else ""

# funcion para escribir parametros de nombre de la tabla
def _build_title(title):
    return ['    TITLE "' + title + '"'] if title else []

# funcion para escribir parametros de nivel de salida de la consulta (areabreak)
def _build_area_break(area_break):
    return "    AREABREAK {}".format(", ".join(area_break)) if area_break else ""

# funcion para escribir parametros de nivel de salida de la consulta (selection)
def _build_selection_inline(selection):
    return "    SELECTION INLINE, {}".format(", ".join(selection)) if selection else ""

# funcion para escribir parametros de variables de la consulta (variables)
def _build_of_variables(variables1):
    if type(variables1) != list:
        variables1 = [variables1]
    return "    OF {}".format(", ".join(filter(lambda x: str(x) if x is not None else '',variables1))) if variables1 else ""

def _build_of_variables2(variables1,variables2):
    if type(variables1) != list:
        variables1 = [variables1]
    if type(variables2) != list:
        variables2 = [variables2]
    return "    AS CROSSTABS OF {} BY {}".format((", ".join(filter(lambda x: str(x) if x is not None else '',variables1))),(", ".join(filter(lambda x: str(x) if x is not None else '',variables2)))) if variables1 or variables2 else ""

# area = ["Provinci"]
# var1 = ["Poblacio.C5P12"]
# var2 = ["Vivienda.C2P4"]
# selection = ["Provinci 0101","Provinci 0102"]
# filter_a = "Distrito"

# query = query_final(tipo="crosstab",variables1=var1,variables2=var2,area_break=area)
# print(query)
# head = "https://censos2017.inei.gob.pe/bininei/RpWebStats.exe/CmdSet?MAIN=WebServerMain.inl&BASE=CPV2017DI&LANG=esp&CODIGO=XXUSUARIOXX&ITEM=PROGRED&MODE=RUN&CMDSET="
# submit = "&Submit=Ejecutar"
# print(head+urllib.parse.quote(query)+submit)
```
